```python
import math
import random
import matplotlib.pyplot as plt
n = [3,4,5,6]
for j in n:
    circle = 0
    square = 0
    circles = []
    squares = []
    for i in range(10**j):
        x = random.uniform(0,1)
        y = random.uniform(0,1)
        r = math.sqrt(x**2 + y**2)
        if r < 1:
            circle += 1
            circles.append((x,y))
        else:
            squares.append((x,y))
        square += 1
    pi = (circle/square)*4
    error = pi/math.pi
    txt = "{result:.6f}"
    print('n = 10 ^ ' + str(j) + ' pi = {0:.6f} error = {1:.4f}%'.format(pi, error))
    zip_circles = zip(*circles)
    plt.scatter(*zip_circles, marker='+', color='b', label='Circle')
    zip_squares = zip(*squares)
    plt.scatter(*zip_squares, marker='+', color='r', label='Square')
    plt.legend()
    plt.show()
```
